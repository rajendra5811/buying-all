# Create a DataFrame -df_src

```python
df_src = spark.sql("""
select distinct CustomerID as CustomerName, Country from parquet.`abfss://silver@cf22acc.dfs.core.windows.net/silverdata`
""")
df_src.display()
```

# Getting Gold Layer Data -df_sink

```python
if spark.catalog.tableExists("gold.dim_customer"):
  df_sink = spark.sql(
    """
    select Dim_Customer_Key,CustomerID,CustomerName,Country from parquet.`abfss://gold@cf22acc.dfs.core.windows.net/dim_customer`
    """
)
else:
  df_sink =spark.sql(
    """
    select 1 as Dim_Customer_Key,CustomerID,CustomerName,Country from parquet.`abfss://silver@cf22acc.dfs.core.windows.net/silverdata` where 1=0
    """
  )
#df_sink = spark.createDataFrame([],schema="Dim_Customer_Key int,CustomerID int,CustomerName string,Country string")
df_sink.display()
```

# Old and new Records - left join

```python

df=df_src.join(df_sink,df_src['CustomerID']==df_sink['CustomerId'],'left')
df.display()
```
